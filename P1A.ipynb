```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
def direct_pi(N):
  N_hits = 0
  for i in range(N):
    x, y = np.random.uniform(-1, 1), np.random.uniform(-1, 1)
    if x**2 + y**2 <= 1:
      N_hits += 1
  return N_hits / N


N_values = [10, 100, 1000, 10000, 100000, 1000000, 10000000, 100000000]
runs = 20
pi_estimates = []
deviations = []
```

```python
for N in N_values:
  temp_deviations = []
  for i in range(runs):
    pi_estimate = direct_pi(N)
    pi_estimates.append(pi_estimates)
    deviation = (pi_estimate - np.pi/4)**2
    temp_deviations.append(deviation)
  deviations.append(np.sqrt(np.mean(temp_deviations)))
```

```python
plt.figure(figsize=(8, 6))
plt.plot(N_values, deviations, marker='o', linestyle='-', color='purple')
plt.xscale('log')
plt.yscale('log')
plt.xlabel('N (número de lanzamientos)')
plt.ylabel('Desviación cuadrática media')
plt.title('Desviación cuadrática media en función de N - Direct-Pi')
plt.grid(True)
plt.show()
```

```python
for i, N in enumerate(N_values):
    print(f"N = {N}, Desviación cuadrática media: {deviations[i]}")
```
